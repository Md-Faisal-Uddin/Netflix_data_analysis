# **Top 10 Director**

```python
df
```

```python
df['director']
```

```python
#fill director nan value by 'No Specific Director'
df['director'] = df['director'].fillna('No specified Director')
df
```

```python
#dataframe
#split director name by coma ,expand it and stack
#convert to data frame
#rename column

filtered_director = pd.DataFrame()
filtered_director = df['director'].str.split(',',expand = True).stack()
filtered_director = filtered_director.to_frame()
filtered_director.columns = ['Director']
filtered_director
```

```python
#groupby Director and name the column Total movie

directors = filtered_director.groupby('Director').size().reset_index(name = 'Total content')
directors
```

```python
#sort directors by Total movie number

directors = directors.sort_values(by = ['Total content'], ascending = False)
directors
```

```python
#remove No specified director
#take top 10 directors
directors = directors[directors.Director != 'No specified Director']
directors = directors.head(10)
directors
```

```python
fig = px.bar(directors, x="Total content", y="Director", orientation='h')
fig.show()
```

```python
import plotly.express as px
fig = px.pie(directors, values='Total content', names='Director')
fig.show()
```

# **Top 10 cast**

```python
df
```

```python
df['cast']
```

```python
#fill cast nan value by 'No Specific Cast'
df['cast'] = df['cast'].fillna('No specified Cast')
df
```

```python
#dataframe
#split cast name by coma ,expand it and stack
#convert to data frame
#rename column

filtered_cast = pd.DataFrame()
filtered_cast = df['cast'].str.split(',',expand = True).stack()
filtered_cast = filtered_cast.to_frame()
filtered_cast.columns = ['Cast']
filtered_cast
```

```python
#groupby Cast and name the column Total movie

casts = filtered_cast.groupby('Cast').size().reset_index(name = 'Total content')
casts
```

```python
#sort casts by Total movie number

casts = casts.sort_values(by = ['Total content'], ascending = False)
casts
```

```python
#remove No specified cast
#take top 10 casts
casts = casts[casts.Cast != 'No specified Cast']
casts = casts.head(10)
casts
```

```python
fig = px.bar(casts, x="Total content", y="Cast", orientation='h')
fig.show()
```

```python
import plotly.express as px
fig = px.pie(casts, values='Total content', names='Cast')
fig.show()
```
